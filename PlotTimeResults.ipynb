```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
inputFile = "time_results.txt"
data = pd.read_table(inputFile, delim_whitespace=True)
data.sort_values(by='n', inplace=True, ignore_index=True)
data
```

```python
plt.plot(data['n']**2,data['time'], marker="x", color='orange')
plt.xlabel("$n^2$")
plt.ylabel("Time [seconds]")
plt.grid()
plt.show()
```

```python
plt.plot(data['n'],data['time'], marker="x", color='orange')
plt.xlabel("$n$")
plt.ylabel("Time [seconds]")
plt.grid()
plt.show()
```
